```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
 def linear_convection(nx, L=2.0, c=1.0, dt=0.025, nt=20):
    dx = L / (nx - 1)
    x = numpy.linspace(0.0, L, num=nx)
    u0 = numpy.ones(nx)
    mask = numpy.where(numpy.logical_and(x >= 0.5, x <= 1.0))
    u0[mask] = 2.0
    u = u0.copy()
    for n in range(1, nt):
        u[1:] = u[1:] - c * dt / dx * (u[1:] - u[:-1])
    pyplot.figure()
    pyplot.xlabel('x')
    pyplot.ylabel('u')
    pyplot.grid()
    pyplot.plot(x, u0, label='Initial')
    pyplot.plot(x, u, label='nt = {}')
    pyplot.legend()
    pyplot.xlim(0.0, L)
    pyplot.ylim(0.0, 2.5);
```

```python
linear_convection(41)
```

```python
linear_convection(81)
```

```python
 def linear_convection_cfl(nx, L=2.0, c=1.0, nt=20,sigma = 0.5):
    dx = L / (nx - 1)
    x = numpy.linspace(0.0, L, num=nx)
    dt = sigma * dx / c
    u0 = numpy.ones(nx)
    mask = numpy.where(numpy.logical_and(x >= 0.5, x <= 1.0))
    u0[mask] = 2.0
    u = u0.copy()
    for n in range(1, nt):
        u[1:] = u[1:] - c * dt / dx * (u[1:] - u[:-1])
    pyplot.figure()
    pyplot.xlabel('x')
    pyplot.ylabel('u')
    pyplot.grid()
    pyplot.plot(x, u0, label='Initial')
    pyplot.plot(x, u, label='nt = {}')
    pyplot.legend()
    pyplot.xlim(0.0, L)
    pyplot.ylim(0.0, 2.5);
```

```python
linear_convection_cfl(85, sigma = 1.1)
```

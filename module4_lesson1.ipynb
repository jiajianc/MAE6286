```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

```python
L = 1.0  
nx = 51  
dx = L / (nx - 1)  
alpha = 1.22e-3 
x = numpy.linspace(0.0, L, num=nx)
T0 = numpy.zeros(nx)
T0[0] = 100.0
sigma = 0.5
dt = sigma * dx**2/alpha
mt =100
```

```python
def FTCS(T0, alpha, dx, dt, mt):
    T = T0.copy()
    for i in range (mt):
        T[1:-1] = T[1:-1] + sigma * (T[2:] - 2*T[1:-1]+T[:-2])
    return T
```

```python
def plot(x, T):
    pyplot.figure()
    pyplot.grid()
    pyplot.xlabel('Distance [m]')
    pyplot.ylabel('Temperature [C]')
    pyplot.plot(x, T)   
    pyplot.show
T = FTCS(T0, alpha, dx, dt, mt)
```

```python
plot(x, T)
```

```python
T_1000 = FTCS(T0, alpha, dx, dt, 1000)
```

```python
plot(x, T_1000)
```

```python
def FTCS2(T0, alpha, dx, dt, mt, q = 0):
    T = T0.copy()
    for i in range (mt):
        T[1:-1] = T[1:-1] + sigma * (T[2:] - 2*T[1:-1]+T[:-2])
        T[-1] = T[-2] + q * dx
    return T
```

```python
T_2 = FTCS2(T0, alpha, dx, dt, mt, q = 0)
plot(x, T_2)
```

```python
T_2_1000 = FTCS2(T0, alpha, dx, dt, 1000, q = 0)
plot(x, T_2_1000)
```
